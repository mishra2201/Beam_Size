```python
#plotting transition rate for each ion while varying the axial position of the beam  

beam_apositions=np.array([149.5, 150.5,151,151.5,152,152.5,153])
rate=[]

for i in range (4):
    
    rates_yscan=np.array([rates_yscan_149s[i],rates_yscan_150s[i],rates_yscan_151[i],rates_yscan_151s[i],rates_yscan_152[i],rates_yscan_152s[i],rates_yscan_153[i]])
    rate.append(rates_yscan) 
    print (rates_yscan)
    fig, ax = plt.subplots(1, figsize = (5, 5))
    ax.scatter(beam_apositions, rates_yscan)
    ax.set_xlabel("Beam Axial Position")
    ax.set_ylabel("Transition Rate")
    ax.set_title(f'Ion {i+1}')
    
print (rates_yscan_149s,"149"),
print (rate)
```
